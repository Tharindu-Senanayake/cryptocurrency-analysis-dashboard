# Crypto Analysis
Load and clean the daily snapshot data.

```python
from pathlib import Path
import pandas as pd

# Resolve project root regardless of where Jupyter sets the working directory
_cwd = Path().resolve()
BASE_DIR = _cwd if (_cwd / 'output').exists() else _cwd.parent

df = pd.read_csv(BASE_DIR / 'output/raw/crypto_daily_snapshots.csv', encoding='utf-8-sig')
df
```

```python
df = df.drop_duplicates(subset=['timestamp', 'symbol'], keep = 'first')
```

```python
df
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
df
```

```python
print(df[['timestamp', 'last_updated', 'aud_last_updated']].dtypes)
```

```python
df2 = df[['timestamp', 'id', 'symbol', 'name', 'slug', 'cmc_rank', 'date_added', 
        'max_supply', 'circulating_supply', 'total_supply', 'infinite_supply', 
        'last_updated', 'aud_price','aud_volume_24h', 
       'aud_cex_volume_24h', 'aud_dex_volume_24h','aud_volume_change_24h', 
       'aud_percent_change_1h', 'aud_percent_change_24h', 'aud_percent_change_7d',
       'aud_percent_change_30d', 'aud_percent_change_60d', 'aud_percent_change_90d', 
       'aud_market_cap', 'aud_market_cap_dominance', 'aud_fully_diluted_market_cap', 
       'global_eth_dominance', 'global_btc_dominance', 'global_defi_volume_24h', 
       'global_defi_market_cap', 'global_stablecoin_volume_24h', 
       'global_stablecoin_market_cap', 'global_derivatives_volume_24h']].copy()
```

```python
len(df2.columns)
```

```python
df2['timestamp'] = pd.to_datetime(df2['timestamp'])
df2['last_updated'] = pd.to_datetime(df2['last_updated'])
df2['date_added'] = pd.to_datetime(df2['date_added'])

print(df2[['timestamp', 'last_updated', 'date_added']].dtypes)
```

```python
print(df2.dtypes)
```

```python
df2['id'] = df2['id'].astype(int)
df2['cmc_rank'] = df2['cmc_rank'].astype(int)
```

```python
df2['date'] = pd.to_datetime(df2['timestamp']).dt.normalize()
print(df2['date'].dtype)
```

```python
dim_coin = df2[['id', 'symbol', 'name', 'slug', 'date_added',  'max_supply', 'infinite_supply']].copy()
dim_coin = dim_coin.drop_duplicates(subset=['id'], keep='first')
```

```python
dim_coin
```

```python
fact_coin_prices = df2[['timestamp', 'date', 'id', 'cmc_rank', 'circulating_supply', 
                       'total_supply', 'aud_price', 'aud_volume_24h', 'aud_cex_volume_24h',
                       'aud_dex_volume_24h', 'aud_volume_change_24h', 'aud_percent_change_1h', 'aud_percent_change_24h',
                       'aud_percent_change_7d', 'aud_percent_change_30d', 'aud_percent_change_60d',
                       'aud_percent_change_90d', 'aud_market_cap', 'aud_market_cap_dominance',
                       'aud_fully_diluted_market_cap']].copy()
fact_coin_prices
```

```python
fact_global_market = df2[['timestamp', 'date', 'global_btc_dominance', 'global_eth_dominance',
                          'global_defi_volume_24h', 'global_defi_market_cap',
                          'global_stablecoin_volume_24h', 'global_stablecoin_market_cap',
                          'global_derivatives_volume_24h']]
fact_global_market = fact_global_market.drop_duplicates(subset=['timestamp', 'date'])
fact_global_market
```

```python
dim_date = df2[['date']]
dim_date = dim_date.drop_duplicates()
dim_date['year'] = dim_date['date'].dt.year
dim_date['quarter'] = dim_date['date'].dt.quarter
dim_date['month'] = dim_date['date'].dt.month
dim_date['day'] = dim_date['date'].dt.day
dim_date['month_name'] = dim_date['date'].dt.month_name()
dim_date['week'] = dim_date['date'].dt.isocalendar().week
dim_date['day_of_week'] = dim_date['date'].dt.dayofweek
dim_date['day_name'] = dim_date['date'].dt.day_name()
```

```python
dim_date
```

```python
(BASE_DIR / 'output/tables').mkdir(parents=True, exist_ok=True)

dim_coin.to_csv(BASE_DIR / 'output/tables/dim_coin.csv', index=False, encoding='utf-8-sig', float_format='%.5f')
dim_date.to_csv(BASE_DIR / 'output/tables/dim_date.csv', index=False, encoding='utf-8-sig', float_format='%.5f')
fact_coin_prices.to_csv(BASE_DIR / 'output/tables/fact_coin_prices.csv', index=False, encoding='utf-8-sig', float_format='%.5f')
fact_global_market.to_csv(BASE_DIR / 'output/tables/fact_global_market.csv', index=False, encoding='utf-8-sig', float_format='%.5f')
```
